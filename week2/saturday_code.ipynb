Colleting data on web services with using BeatifulSoup Libary 

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://kararlarbilgibankasi.anayasa.gov.tr/BB/2019/1628'
r = requests.get(url)
r.text
bs = BeautifulSoup(r.text) # texti uraşabilmesi ise uygun hale getirmek için bs4 kullandık.
raw_data = bs.find('div',{'id':'Karar'}).text
raw_data.replace("\n"," ").replace("\xa0"," ")
```

```python
table_responsive = bs.find('div',{'class':'table-responsive'})
table_responsive.table.tbody.find_all('td')[2].text.replace(" ","").replace("\n","")
```

```python
data_text = list()
for year in range(2020,2022):
    for case_no in range(1600,7200):
        url = f'https://kararlarbilgibankasi.anayasa.gov.tr/BB/{year}/{case_no}'
        r = requests.get(url)
        bs = BeautifulSoup(r.text)
        raw_data = bs.find('div',{'class':'WordSection1'})
        if raw_data is None:
            continue
        raw_data = raw_data.text
        result = raw_data.replace("\n"," ").replace("\xa0"," ")
        table_responsive = bs.find('div',{'class':'table-responsive'})
        result_1 = table_responsive.table.tbody.find_all('td')[2].text.replace(" ","").replace("\n","")
        
        data_text.append(
            {
                'sonuc':result_1,
                'dava': result
            }
        )
data_text
```

```python
for i in data_text:
    case = i['dava']
    start_index = case.find('I. BAŞVURUNUN KONUSU')
    stop_index = case.find('IV. İNCELEME VE GEREKÇE')
    i['dava']=case[start_index:stop_index]
data_text
```

```python
import pandas as pd

dataframe = pd.DataFrame(data_text)
dataframe.to_csv('caseno_result.csv')
```

```python
import datetime
```

```python
def return_dates(start,end):
    time_list=[start.strftime("%Y-%m-%d")]
    while start != end:
        start += datetime.timedelta(days=1)
        time_list.append(start.strftime("%Y-%m-%d"))
    return time_list    
```
