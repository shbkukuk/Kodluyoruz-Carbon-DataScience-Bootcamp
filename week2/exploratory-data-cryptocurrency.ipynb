Comparing change volume of last 24 hours how effect on changes value of last 24 hours

```python
own_data = dataframe[['Crypto Name','Changes 24H ','Volume 24H']]
own_data
```

All columns is object, That's so we convert to number.

```python
own_data['Changes 24H ']=own_data['Changes 24H '].str.replace('%','')
own_data['Changes 24H ']=own_data['Changes 24H '].str.replace('$','')
own_data['Changes 24H ']=own_data['Changes 24H '].astype(float)

own_data
```

Volume 24H columns contain shorting of Millionare, Billionare and 'K' stands for 'kilo. The code cell is converted to 'B' -> 10^9, 'M' -> 10^6, K -> 10^3.

```python
own_data['Volume 24H']=own_data['Volume 24H'].str.replace('$','')
for i in range(len(own_data['Volume 24H'])):
    if own_data['Volume 24H'][i] is not float:
        price_list = own_data['Volume 24H'][i].split(" ")
        price = 0
        if price_list[1]== 'B':
            price = float(price_list[0])*(10**9)
        elif price_list[1]== 'M':
            price = float(price_list[0])*(10**6)
        elif price_list[1]== 'K':
            price = float(price_list[0])*(10**3)
        own_data['Volume 24H'][i] = price
        

own_data

```

```python
own_data['Volume 24H']=own_data['Volume 24H'].astype(float)
own_data.info()
```
